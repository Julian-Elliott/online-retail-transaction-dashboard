# **Cleaned Retail Dataset with Histograms**

## Objectives

* Load the cleaned Online Retail Transaction dataset and perform exploratory data visualisation by generating histograms for all numeric columns.

* Identify the distribution patterns and outliers in key metrics such as Quantity, UnitPrice, and TotalPrice to guide future analysis and dashboard design.


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load the cleaned dataset
df = pd.read_csv('../data/clean_online_retail.csv')

# Show top 5 rows to verify the load
df.head()
```

```python
# Show data types and nulls
df.info()
```


## Outputs

* Visual outputs: Multi-panel histogram plots of numeric columns


```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histograms for all numeric columns in a multi-plot layout
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()

plt.suptitle(" Histograms of Numeric Columns", fontsize=16, y=1.05)
plt.show()
```


## Additional Comments

* Data cleaning (removing nulls, handling cancellations, and computing TotalPrice) has already been completed in a prior ETL step.

* Histogram plots help confirm data integrity and guide decisions about scaling, segmentation, or transformations required for business analysis in Power BI/Tableau.
